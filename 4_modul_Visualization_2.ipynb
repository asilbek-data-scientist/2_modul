```python
import matplotlib.pyplot as plt, pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df = df[df.year>=2010]
df = df[df.transmission!="Other"]
df =df[df.fuelType!='Other']
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(20,10))
sns.histplot(data=df, x='price')
plt.title("Merc")
plt.grid(linestyle="--")
```

```python
df = df[df.price<100000]
sns.displot(data=df, x='price', hue='transmission', multiple="stack", height=10, aspect=2) #hue navbatdagi ustun qo'shish 
plt.grid(linestyle="--")
plt.show()
```

```python
df = df[df.price<100000]
sns.displot(data=df, x='price', col='transmission', multiple="stack", height=10, aspect=2) #col alohida jadval ko'rinishida chiqaradi. 
plt.grid(linestyle="--")
plt.show()
```

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', kde=True, height=5, aspect=2)
plt.grid(linestyle="--")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', shade=True)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.kdeplot(data=df, x="price", hue="fuelType", shade=True)
plt.grid()
plt.show()
```

```python
sns.displot(data=df, x='mileage', kind='ecdf', height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which="major") # setka (major, minor, both) qiymalarini berib ko'ring
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='price', kind='ecdf', height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which="both") # setka (major, minor, both) qiymalarini berib ko'ring
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='price', hue="fuelType", kind='ecdf', height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which="both") # setka (major, minor, both) qiymalarini berib ko'ring
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='price', y='year', hue="fuelType", kind='kde', height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which="both") # setka (major, minor, both) qiymalarini berib ko'ring
plt.show()
```

#Subplot

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('http://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df = df[df.year>=2010]
df = df[df.transmission!='Other']
df = df[df.fuelType!="Other"]
df.head()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos =  np.cos(x)

plt.figure(figsize=(12, 6))

#sin(x) grafigi
plt.subplot(2, 1, 1)
plt.plot(x, sin,
          linewidth=1.5,
          linestyle="-.",
          color="#8E44AD",
          alpha=0.7,
          marker="h")
plt.title("sin(x) funksiya grafigi")
plt.grid(color='tomato', alpha=0.5, linestyle="--")

#cos(x) grafigi
plt.subplot(2, 1, 2)
plt.plot(x, cos,
          linewidth=1.5,
          linestyle=":",
          color="c",
          alpha=0.7,
          marker="v")
plt.title("cos(x) funksiya grafigi")
plt.grid(color='tomato', alpha=0.5, linestyle="--")
plt.show()
```

```python
sns.histplot(data=df, x='price')
plt.show()
```

```python
sns.scatterplot(data=df, x='price', y='year', s=100)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
sns.histplot(ax=ax[0], data=df, x='price')
sns.scatterplot(ax=ax[1], data=df, x='price', y='year', s=100)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(25,15))
# Umumiy chizma nomini beramiz:
fig.suptitle("Grafik 2")
sns.histplot(ax=ax[0, 0], data=df, x='price')
sns.histplot(ax=ax[0, 1], data=df, x='mileage')
sns.scatterplot(ax=ax[1, 0], data=df, x=df['year'], y=df['price'], hue="fuelType", s=50)
sns.scatterplot(ax=ax[1, 1], data=df, x=df['year'], y=df['mileage'], hue="fuelType", s=50)
# Har bir grafik uchun nom:
ax[0,0].set_title("Price distribution")
ax[0,1].set_title("Mileage distribution")
ax[1,0].set_title("Price vs Year correlation")
ax[1,1].set_title("Mileage vs Year correlation")
plt.show()
```
