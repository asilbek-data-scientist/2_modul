```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
plt.plot(x, y, linewidth=2, linestyle=":", color="r", marker=">")
plt.title("sin(x) funksiya")
plt.xlabel("PI")
plt.ylabel("cos(x)")
plt.grid(color='g', linestyle="--")
plt.show()
```

```python
x = np.arange(0,4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12, 6))
#sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle="-.",
         color="#8E44AD",
         alpha=0.7,
         marker='h')
#cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=":",
         color="c",
         alpha=0.7,
         marker="v")
plt.title("Sin(x) va cos(x) grafigi")
plt.legend(["sin(x) funksiya", "cos(x) funksiya"], loc=0, fontsize='x-large')
plt.grid(color="r", linestyle="--")
plt.savefig("grafig1.png")
plt.show()
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv', index_col=0, parse_dates=[0], usecols=[0,1])
df.head()
```

```python
type(df.index)
```

```python
df['month'] = [i.month for i in df.index]
df['year'] = [y.year for y in df.index]
df.head()
```

```python
df2 = df.groupby(['month', 'year']).max()
df2.head()
```

```python
df2.unstack(level=1)
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(data=df2.unstack(level=0), cmap="YlGnBu", linewidths=1, square=True)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', "June", "July",
                 "Aug", 'Sep', 'Oct', 'Now', 'Dec']
plt.xticks(np.arange(12) + 0.5, labels=xticks_labels)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20, 10))
sns.barplot(x=df['Month'], y=df["AA"])
plt.grid(color="black", linestyle="--")
plt.ylabel("Minutes")
plt.title("American Airlines")
plt.show()
```

```python
data=df.melt(["Month"])
data
```

```python
#plt.figure(figsize=(20, 18))
sns.catplot(data=data, kind="bar", x="Month", y='value', hue="variable")
plt.grid(color="black", linestyle="--")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(data=df.set_index('Month'), cmap="Accent", annot=True)
plt.xlabel("Airline")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df.head()
```

```python
plt.figure(figsize=(10, 5))
df = df[df.year>2000]
sns.scatterplot(data=df, x="year", y='price')
plt.grid(linestyle="--")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df = df[df.year>2000]
sns.regplot(data=df, x="year", y='price')
plt.grid(linestyle="--")
plt.show()
```

```python
plt.figure(figsize=(25, 17))
df = df[df.year>2000]
sns.scatterplot(data=df, x="mileage", y='price', hue="fuelType", s=200)
plt.grid(linestyle="--")
plt.show()
```

```python
sns.swarmplot(data=df, x=df['fuelType'], y=df['price'])
plt.show()
```
